```python

import re
import os

articles_dict = {}
folder_articles = "articles"

for filename in os.listdir(folder_articles):
    file_path = os.path.join(folder_articles, filename)

    if os.path.isfile(file_path):
        try:
            with open(file_path, 'r', encoding='utf-8') as file:
                text = file.read().strip()  # read whole file, remove leading/trailing whitespace

                # Use filename without extension as string key
                key = os.path.splitext(filename)[0]  # this is safe for filenames like 'Netclan20241017.txt'
                articles_dict[key] = text

        except Exception as e:
            print(f"Exception '{e}' occurred at file {filename}")
            key = os.path.splitext(filename)[0]
            articles_dict[key] = ''
```

```python
import re

def extract_words(text_data):
    text = re.sub(r"[^a-zA-Z0-9]", " ", text_data.lower())
    words = text.split()
    specials = r'@_!#$%^&*()<>?/\|}{~:;.[]'  # raw string
    words_list = [''.join(i for i in word if i not in specials) for word in words]
    return words_list
```

```python
import os
stopwords = []
folder_StopWords = "/content/drive/MyDrive/StopWords"
for filename in os.listdir(folder_StopWords):
    file_path = os.path.join(folder_StopWords, filename)

    if os.path.isfile(file_path):
        with open(file_path, 'r', encoding='latin-1') as file:
            result = file.read()
            stopwords += extract_words(result)
```

```python
clean_articles = {}
for ix, text in articles_dict.items():
    words = extract_words(text)
    words = [word for word in words if word not in stopwords]
    clean_articles[ix] = words
```

```python
master_dict = {}
folder_masterD = "/content/drive/MyDrive/MasterDictionary"
for filename in os.listdir(folder_masterD):
    file_path = os.path.join(folder_masterD, filename)

    if os.path.isfile(file_path):
        with open(file_path, 'r', encoding='latin-1') as file:
            result = file.read()
            master_dict[filename[:8]] = result.split('\n')
```

```python
def positive_score(words):
    score=0
    for word in words:
        if word in master_dict['positive']:
            score += 1
    return score

def negative_score(words):
    score=0
    for word in words:
        if word in master_dict['negative']:
            score += 1
    return score

def polarity_score(p_score, n_score):
    score = (p_score - n_score)/ ((p_score + n_score) + 0.000001)
    return score

def subjectivity_score(p_score, n_score, n):
    score = (p_score + n_score)/ (n + 0.000001)
    return score
```

```python
def count_syllables(text):
    vowels = "aeiouy"

    words = text.split()
    syllable_counts = []
    complex_word_count = 0

    for word in words:
        syllables = 0
        prev_char = ''

        for char in word.lower():
            if char in vowels and prev_char not in vowels:
                syllables += 1
            prev_char = char

        if word.lower().endswith(("es", "ed")):
            syllables -= 1

        if syllables == 0:
            syllables = 1

        syllable_counts.append(syllables)
        if syllables > 2:
            complex_word_count += 1

    return syllable_counts, complex_word_count

def count_personal_pronouns(text):
    pronouns = r"\b(I|we|my|ours|us)\b"
    matches = re.findall(pronouns, text, flags=re.IGNORECASE)
    US_count = re.findall(r'\b(US)\b', text)
    return len(matches) - len(US_count)
```

```python
import os
import nltk
nltk.download('stopwords')

# Prepare stopwords once
stop_words = set(nltk.corpus.stopwords.words('english'))

# Load articles safely
articles_dict = {}
folder_articles = "articles"

for filename in os.listdir(folder_articles):
    file_path = os.path.join(folder_articles, filename)
    if os.path.isfile(file_path):
        try:
            with open(file_path, 'r', encoding='utf-8') as file:
                text = file.read().strip()
                key = os.path.splitext(filename)[0]  # Keep as string
                articles_dict[key] = text
        except Exception as e:
            print(f"Exception '{e}' occurred at file {filename}")
            key = os.path.splitext(filename)[0]
            articles_dict[key] = ''

# Ensure URL_ID in output is string to match dictionary keys
output['URL_ID'] = output['URL_ID'].astype(str)

# Process articles and update DataFrame
for ix, article_text in articles_dict.items():
    # Extract clean words for scoring if you have a clean_articles dict
    clean_words = clean_articles.get(ix, [])  # empty list if not found

    # Syllables and complex words
    syllable_counts, complex_word_count = count_syllables(article_text)

    words_in_article = extract_words(article_text)
    num_of_words = len(words_in_article)

    p_score = positive_score(clean_words)
    n_score = negative_score(clean_words)
    n = len(clean_words)

    num_sentences = article_text.count('.') if article_text.count('.') else 1

    avg_sen_len = num_of_words / num_sentences
    per_of_com_words = complex_word_count / num_of_words if num_of_words else 0
    fog_index = 0.4 * (avg_sen_len + per_of_com_words)

    word_lengths = [len(word) for word in words_in_article]
    avg_word_length = sum(word_lengths) / num_of_words if num_of_words else 0

    word_count_no_stop = len([word for word in words_in_article if word.lower() not in stop_words])

    personal_pronouns = count_personal_pronouns(article_text)

    # Update DataFrame safely
    mask = output['URL_ID'] == ix
    if mask.any():  # Only update if row exists
        output.loc[mask, 'POSITIVE SCORE'] = p_score
        output.loc[mask, 'NEGATIVE SCORE'] = n_score
        output.loc[mask, 'POLARITY SCORE'] = polarity_score(p_score, n_score)
        output.loc[mask, 'SUBJECTIVITY SCORE'] = subjectivity_score(p_score, n_score, n)
        output.loc[mask, 'AVG SENTENCE LENGTH'] = avg_sen_len
        output.loc[mask, 'PERCENTAGE OF COMPLEX WORDS'] = per_of_com_words
        output.loc[mask, 'FOG INDEX'] = fog_index
        output.loc[mask, 'AVG NUMBER OF WORDS PER SENTENCE'] = avg_sen_len
        output.loc[mask, 'COMPLEX WORD COUNT'] = complex_word_count
        output.loc[mask, 'WORD COUNT'] = word_count_no_stop
        output.loc[mask, 'SYLLABLE PER WORD'] = sum(syllable_counts)
        output.loc[mask, 'PERSONAL PRONOUNS'] = personal_pronouns
        output.loc[mask, 'AVG WORD LENGTH'] = avg_word_length
    else:
        print(f"Warning: URL_ID '{ix}' not found in output DataFrame")
```
